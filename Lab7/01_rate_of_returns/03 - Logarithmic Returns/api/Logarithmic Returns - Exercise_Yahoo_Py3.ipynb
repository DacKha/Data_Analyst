## Logarithmic Returns

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

MSFT = yf.download('MSFT', start='2015-01-01')
print(MSFT.head())
```

### Log Returns

$$
ln(\frac{P_t}{P_{t-1}})
$$

Calculate the Log returns for Microsoft.

```python
MSFT['log-return']= np.log(MSFT['Close'] / MSFT['Close'].shift(1))
MSFT['log-return']
```

Plot the results on a graph.

```python
MSFT['log-return'].plot(figsize=(8, 5))
plt.show()
```

Estimate the daily and the annual mean of the obtained log returns.

```python
log_return_d = MSFT['log-return'].mean()
log_return_d
```

```python
log_return_a = MSFT['log-return'].mean() * 250
log_return_a
```

Print the result in a presentable form.

```python
print (str(round(log_return_a, 5) * 100) + ' %')
```
