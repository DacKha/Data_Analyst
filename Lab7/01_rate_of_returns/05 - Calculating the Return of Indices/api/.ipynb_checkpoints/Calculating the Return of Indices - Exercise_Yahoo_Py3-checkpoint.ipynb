## Calculating the Return of Indices

Consider three famous American market indices â€“ Dow Jones, S&P 500, and the Nasdaq for the period of 1st of January 2000 until today.

```python
import pandas as pd
import yfinance as yf

data = yf.download(['^DJI', '^GSPC', '^IXIC'], start='2000-01-01', end='2025-11-01')
adj_close = data['Close'].copy()
adj_close.head()
```

Normalize the data to 100 and plot the results on a graph. 

```python
normalized = adj_close / adj_close.iloc[0] * 100
```

```python
import matplotlib.pyplot as plt
normalized.plot(figsize=(15, 6));
plt.show()
```

Obtain the simple returns of the indices.

```python
adj_close['DJI_simple_return'] = adj_close['^DJI'].pct_change()
adj_close['GSPC_simple_return'] = adj_close['^GSPC'].pct_change()
adj_close['IXIC_simple_return'] = adj_close['^IXIC'].pct_change()
print(adj_close[['Date', 'DJI_simple_return', 'GSPC_simple_return', 'IXIC_simple_return']].head())
```

Estimate the average annual return of each index.

```python
mean_daily = adj_close[['DJI_simple_return', 'GSPC_simple_return', 'IXIC_simple_return']].mean()

annual_returns = (1 + mean_daily) ** 250 - 1

print("Average Annual Return (2000â€“2025):")
print(annual_returns)
```
