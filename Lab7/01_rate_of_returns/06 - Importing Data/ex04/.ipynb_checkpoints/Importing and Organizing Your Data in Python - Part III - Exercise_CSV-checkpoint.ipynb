```python
import pandas as pd 
import yfinance as yf
```

```python
tickers = ['AAPL', 'MSFT', 'XOM', 'BP']
new_data = pd.DataFrame()
for t in tickers:
    data = yf.download(t, start='2015-01-01', progress=False)
    new_data[t] = data['Close']
```

```python
new_data.to_csv('example_complete_1.csv')
```

```python
new_data.to_excel('example_complete_1.xlsx')
```
