## Simple Returns - Part II

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
MSFT = yf.download('MSFT', start='2015-01-01', progress=False)

MSFT['simple_return'] = (MSFT['Close'] / MSFT['Close'].shift(1)) - 1

print("Dữ liệu MSFT:")
print(MSFT.tail())
print("\n" + "="*50 + "\n")
print("Simple Return:")
print(MSFT['simple_return'])
```

Plot the simple returns on a graph.

```python
plt.figure(figsize=(14, 6))
plt.plot(MSFT.index, MSFT['simple_return'], linewidth=0.8, color='blue', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=1, alpha=0.5)
plt.title('MSFT Simple Returns (2015 - Present)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Simple Return', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(MSFT['simple_return'].dropna(), bins=100, edgecolor='black', alpha=0.7, color='skyblue')
plt.axvline(x=0, color='red', linestyle='--', linewidth=2)
plt.title('Distribution of MSFT Simple Returns', fontsize=16, fontweight='bold')
plt.xlabel('Simple Return', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Calculate the average daily return.

```python
avg_daily_return = MSFT['simple_return'].mean()
avg_daily_return
```

Estimate the average annual return.

```python
avg_returns_a = MSFT['simple_return'].mean() * 250
avg_returns_a
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print (str(round(avg_returns_a, 4) * 100) + ' %')
```
