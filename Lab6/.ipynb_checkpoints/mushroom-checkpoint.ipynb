Những Quan Sát Chính Những điều cần quan sát

Tách Biệt Lớp: Các danh mục trong mỗi đặc điểm có độc quyền cho một lớp không
Sự Chiếm Ưu: Các danh mục trong một đặc điểm cụ thể có chiếm ưu thế trong một lớp không, ví dụ như ăn được hoặc độc hại
Sự Chồng Chéo: Các danh mục được chia sẻ giữa hai lớp.
Ví dụ: 'Mùi' so với 'lớp'

Tách Biệt Lớp Hoàn Hảo: Nhiều danh mục mùi độc quyền cho một lớp, khiến mùi trở thành một đặc điểm dự đoán cao: a (hạnh nhân), l (cá hồi), và n (không có) chỉ xuất hiện trong nấm ăn được c (creosote), f (hôi), m (ẩm mốc), p (nồng), s (cay), và y (cá) chỉ xuất hiện trong nấm độc
Mùi Chiếm Ưu: Trong số nấm ăn được, n (không có) chiếm ưu thế với 3408 mẫu. Trong số nấm độc, f (hôi) là phổ biến nhất với 2160 mẫu
Không Có Sự Chồng Chéo: Không có danh mục mùi nào được chia sẻ giữa hai lớp, điều này ngụ ý không có sự mơ hồ nào trong đặc điểm này
'mùi' có khả năng dự đoán gần như hoàn hảo.

# 3| Triển khai mô hình

## I. Mã hóa các thuộc tính

### LabelEncoder

```python
# column Encoded With LabelEcoder technique
colum_to_encode=['class','bruises','gill-attachment','gill-spacing','gill-size','stalk-shape']
labelencoder=LabelEncoder()
for col in colum_to_encode:
    df[col]=labelencoder.fit_transform(df[col])
```

```python
df.head(20)
```

```python
from sklearn.preprocessing import OneHotEncoder
#column to implement one hot encoding 
one_hot_encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = one_hot_encoder.fit_transform(df[['ring-number']])
one_hot_columns = [f"ring-number_{cat}" for cat in one_hot_encoder.categories_[0]]
one_hot_df = pd.DataFrame(one_hot_encoded, columns=one_hot_columns)
data = pd.concat([df.reset_index(drop=True), one_hot_df.reset_index(drop=True)], axis=1)
data = data.drop(columns='ring-number')
```

```python
df.head(10)
```

### FeatureHasher

```python
df.columns = df.columns.str.strip()

# Danh sách các cột cần hash
column_hash_encoding = [
    'cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-color',
    'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring',
    'stalk-color-above-ring', 'stalk-color-below-ring',
    'veil-color', 'ring-number','ring-type', 'spore-print-color', 'population', 'habitat'
]

# Số lượng đặc trưng hash cho mỗi cột
n_hash_features = 4

# Danh sách lưu từng DataFrame hashed
hashed_feature_dfs = []

# Hash từng cột
for col in column_hash_encoding:
    dict_col = df[col].astype(str).apply(lambda x: {x: 1}).tolist()
    hasher = FeatureHasher(n_features=n_hash_features, input_type='dict')
    hashed = hasher.transform(dict_col)
    
    # Binarize kết quả hash (0/1)
    binarizer = Binarizer(threshold=0)
    binary_hashed = binarizer.fit_transform(hashed.toarray())
    
    # Tạo DataFrame từ dữ liệu hash
    hashed_df = pd.DataFrame(
        binary_hashed,
        columns=[f"{col}_hash_{i}" for i in range(n_hash_features)]
    )
    hashed_feature_dfs.append(hashed_df)

# Ghép tất cả lại
hashed_all = pd.concat(hashed_feature_dfs, axis=1)

# Loại bỏ các cột gốc và ghép các cột hash vào
data = df.drop(columns=column_hash_encoding).reset_index(drop=True)
data = pd.concat([data, hashed_all], axis=1)
```

```python
data.head()
```

## II. Train-Test Split

```python
x=data.iloc[:,1:-1]
y=data.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## III. Model Naive Bayes

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
y_predict=bnb.predict(x_test)
```

```python
print('Evaluation:\n')
print(f'Acuuracy: {accuracy_score(y_test,y_predict)*100:0.2f}%')
print('\nClassification Report :\n',classification_report(y_test,y_predict))
```

Nhận xét tổng hợp:

Hiệu suất tổng thể cao:
Với accuracy ≈ 93.85%, mô hình BernoulliNB của bạn hoạt động rất tốt, đặc biệt nếu dữ liệu là nhị phân (0/1 features — ví dụ: bag-of-words hoặc one-hot).

Precision và Recall cân bằng:

Lớp 0: recall cao hơn precision → mô hình ít bỏ sót mẫu lớp 0.

Lớp 1: precision cao hơn recall → mô hình ít dự đoán sai lớp 1, nhưng vẫn bỏ sót một số mẫu thật của lớp 1.

F1-score ~ 0.94 cho cả hai lớp:
→ Cho thấy mô hình rất ổn định và cân bằng giữa precision và recall.
Đây là điều tốt, nhất là trong các bài toán phân loại nhị phân như spam, bệnh lý, hay dự đoán khách hàng rời bỏ.

Macro avg và Weighted avg bằng nhau (~0.94):
Điều này cho thấy dữ liệu hai lớp không bị mất cân bằng quá lớn (số lượng mẫu 843 vs 782 là khá đồng đều).

## IV. Ma trận nhầm lẫn

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Tính toán ma trận nhầm lẫn
cm = confusion_matrix(y_test, y_predict)

# Hiển thị ma trận nhầm lẫn bằng Seaborn
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)

plt.title('Confusion Matrix - BernoulliNB')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

Mô hình hoạt động rất tốt:

Hầu hết các mẫu được dự đoán đúng.

Tổng số sai chỉ 100 mẫu trên 1625 (≈6.15%).

Sai sót chính tập trung ở:

60 mẫu lớp 1 bị dự đoán nhầm thành lớp 0 → mô hình hơi “bảo thủ”, có xu hướng đoán 0 khi không chắc chắn.

Điều này làm giảm Recall của lớp 1 (vì bỏ sót một số mẫu thật của lớp 1).

Tùy mục tiêu mà điều chỉnh:

Nếu bạn muốn giảm lỗi bỏ sót (FN) của lớp 1 (ví dụ trong bài toán phát hiện gian lận, phát hiện bệnh…),
→ có thể hạ ngưỡng dự đoán (threshold) xuống một chút.

Nếu mục tiêu là chính xác tuyệt đối khi dự đoán lớp 1 (ví dụ gắn nhãn spam),
→ nên giữ nguyên vì precision của lớp 1 đã rất cao (≈95%).
