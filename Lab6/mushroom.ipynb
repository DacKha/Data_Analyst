### FeatureHasher

```python
df.columns = df.columns.str.strip()

# Danh s√°ch c√°c c·ªôt c·∫ßn hash
column_hash_encoding = [
    'cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-color',
    'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring',
    'stalk-color-above-ring', 'stalk-color-below-ring',
    'veil-color', 'ring-number','ring-type', 'spore-print-color', 'population', 'habitat'
]

# S·ªë l∆∞·ª£ng ƒë·∫∑c tr∆∞ng hash cho m·ªói c·ªôt
n_hash_features = 4

# Danh s√°ch l∆∞u t·ª´ng DataFrame hashed
hashed_feature_dfs = []

# Hash t·ª´ng c·ªôt
for col in column_hash_encoding:
    dict_col = df[col].astype(str).apply(lambda x: {x: 1}).tolist()
    hasher = FeatureHasher(n_features=n_hash_features, input_type='dict')
    hashed = hasher.transform(dict_col)
    
    # Binarize k·∫øt qu·∫£ hash (0/1)
    binarizer = Binarizer(threshold=0)
    binary_hashed = binarizer.fit_transform(hashed.toarray())
    
    # T·∫°o DataFrame t·ª´ d·ªØ li·ªáu hash
    hashed_df = pd.DataFrame(
        binary_hashed,
        columns=[f"{col}_hash_{i}" for i in range(n_hash_features)]
    )
    hashed_feature_dfs.append(hashed_df)

# Gh√©p t·∫•t c·∫£ l·∫°i
hashed_all = pd.concat(hashed_feature_dfs, axis=1)

# Lo·∫°i b·ªè c√°c c·ªôt g·ªëc v√† gh√©p c√°c c·ªôt hash v√†o
data = df.drop(columns=column_hash_encoding).reset_index(drop=True)
data = pd.concat([data, hashed_all], axis=1)
```

```python
y_predict=bnb.predict(x_test)
```

```python
print('Evaluation:\n')
print(f'Acuuracy: {accuracy_score(y_test,y_predict)*100:0.2f}%')
print('\nClassification Report :\n',classification_report(y_test,y_predict))
```
